```python
import pandas as pd
import numpy as np
from glob import glob
import re
import datetime

all_files = pd.DataFrame(glob(r'C:\Users\deepe\OneDrive\Desktop\NLP\2021-20210705T150540Z-001\2021\May\*'),columns=['location'])
```

```python
all_files
```

```python
all_files['data_date'] = all_files['location'].apply(lambda x: re.findall(r'[0-9]{8}',x)[0])
```

```python
all_files
```

```python
all_files['data_date'] = all_files['data_date'].apply(lambda x: datetime.datetime.strptime(x.strip(),'%Y%m%d').date())
```

```python
all_files
```

```python
all_files.sort_values('data_date',inplace=True)
all_files.reset_index(drop=True,inplace=True)
```

```python
all_files_tickers = pd.DataFrame(columns=['ticker','data_date'])
```

```python
for index, row in all_files.iterrows():
    data = pd.read_csv(row['location'])

    data.columns = data.columns.str.lower()
    data.columns = data.columns.str.strip()
    data.columns = data.columns.str.replace(' ','_')

    temp_tickers_df = pd.DataFrame(columns=['ticker'])
    temp_tickers_df['ticker'] = data['ticker'].drop_duplicates().copy()
    temp_tickers_df['data_date'] = row['data_date']

    all_files_tickers = pd.concat([all_files_tickers,temp_tickers_df],axis=0)
    all_files_tickers = all_files_tickers.drop_duplicates(subset=['ticker'])
    
all_files_tickers = all_files_tickers.sort_values('ticker')
all_files_tickers.reset_index(drop=True,inplace=True)
all_files_tickers.dropna(inplace=True)

all_files_tickers['ticker'] = all_files_tickers['ticker'].apply(lambda x: x.split('.NFO')[0])
```
